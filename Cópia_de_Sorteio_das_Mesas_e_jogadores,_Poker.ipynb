```python
import numpy as np
import time

# Criando uma lista com as cartas do 07 ao Ás
cartas = ['07', '08', '09', '10', 'J', 'Q', 'K', 'A']

# Criando uma lista com os 24 jogadores
jogadores = ['Jogador {}'.format(i) for i in range(1, 25)]

# Embaralhando a lista de jogadores
np.random.shuffle(jogadores)

# Dividindo os jogadores em 3 mesas
mesa_1 = jogadores[:8]
mesa_2 = jogadores[8:16]
mesa_3 = jogadores[16:]

# Sorteando as cartas para cada mesa
def sorteio_cartas(mesa):
    return np.random.choice(cartas, size=len(mesa), replace=False)

# Função para imprimir a mesa com pausa
def imprimir_mesa(mesa, cartas_mesa, nome_mesa):
    print(f"\nMesa de {nome_mesa}:")
    for jogador, carta in zip(mesa, cartas_mesa):
        print(f"{jogador}: {carta}")
        time.sleep(1)  # Pausa de 1 segundo entre cada jogador

# Sorteando e mostrando a mesa de Copas
cartas_mesa_1 = sorteio_cartas(mesa_1)
imprimir_mesa(mesa_1, cartas_mesa_1, "Copas")

# Sorteando e mostrando a mesa de Paus
cartas_mesa_2 = sorteio_cartas(mesa_2)
imprimir_mesa(mesa_2, cartas_mesa_2, "Paus")

# Sorteando e mostrando a mesa de Ouro
cartas_mesa_3 = sorteio_cartas(mesa_3)
imprimir_mesa(mesa_3, cartas_mesa_3, "Ouro")
```
