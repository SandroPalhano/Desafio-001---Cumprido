```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios de vendas por mês
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
vendas = [1000, 1500, 1200, 1800, 2000, 2500, 2200, 1900, 1700, 2300, 2800, 3200]

# Criar um DataFrame a partir dos dados
df = pd.DataFrame({'Mês': meses, 'Vendas': vendas})

# Criar um gráfico de linhas
plt.plot(df['Mês'], df['Vendas'])
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.title('Vendas Mensais')
plt.xticks(rotation=45) # Rotacionar os rótulos do eixo x para melhorar a legibilidade
plt.show()
```

```python
import pandas as pd

# Criando um dicionário de listas
dados = {'Nome': ['Alice', 'Bob', 'Charlie'],
         'Idade': [25, 30, 35],
         'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba']}

# Criando um DataFrame a partir do dicionário
df = pd.DataFrame(dados)

# Exibindo o DataFrame
print(df)
```

```python
import pandas as pd

# Criando um DataFrame
dados = {'Nome': ['Alice', 'Bob', 'Charlie'],
         'Idade': [25, 30, 35],
         'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba']}
df = pd.DataFrame(dados)

# Acessando uma linha pelo índice
linha = df.loc[0]
print(linha)

# Acessando várias linhas pelo índice
linhas = df.loc[[0, 2]]
print(linhas)
```
